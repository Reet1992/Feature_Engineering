```python
### Inference : If its a long Gun what is the probability that it will be handgun
```

```python
X = df2[['long_gun']]
X_array = X.values
```

```python
y = df2[['handgun']]
y_array = y.values
```

```python
print(X_array)
print(y_array)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X_array, y_array, test_size = 1/8, random_state = 0)
```

```python
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
y_pred = regressor.predict(x_test)
```

```python
plt.scatter(y_pred, y_test, color = 'yellow')
plt.plot(y_pred, regressor.predict(y_pred), color = 'black')
plt.show()
from sklearn import metrics
accuracy = metrics.r2_score(y_test,y_pred)
print(accuracy)
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test.argmax(axis=1), y_pred.argmax(axis=1))

cm
plt.matshow(cm)
plt.title('Confusion matrix')
plt.colorbar()
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.show()
```
